```
n, capacity = map(int, input().split())
s = [list(map(int, input().split())) for _ in range(n)]
s.sort(key=lambda i: i[0] / i[1])
result = 0.

for i in s[::-1]:
    if capacity - i[1] >= 0:
        capacity -= i[1]
        result += i[0]
    else: 
        if i[0] != 0:
            result += i[0] / i[1] * capacity
            capacity -= capacity
            
print(round(result,3))
```

```
def heapify(sp, i):
    l = 2*i + 1
    r = 2*i + 2
    global long
    heapsize = long
    largest = i
    if l < heapsize and sp[l] > sp[i]:
        if r < heapsize and sp[l] < sp[r]:
            largest = r
        else: largest = l
    if r < heapsize and sp[r] > sp[i] and sp[l] < sp[r]:
        if l < heapsize and sp[l] > sp[r]:
            largest = l
        else: largest = r
    if largest != i:
        sp[i], sp[largest] = sp[largest], sp[i]
        heapify(sp, largest)


def HeapExtractMax(sp):
    sp[0], sp[-1] = sp[-1], sp[0]
    print(sp.pop())
    global long
    long -= 1
    heapify(sp, 0)


def siftUp(a, i):
    while a[i] > a[(i - 1) // 2] and i > 0:
        a[i], a[(i - 1) // 2] = a[(i - 1) // 2], a[i]
        i = (i - 1) // 2


heap = []
long = 0
quantity_command = int(input())
while quantity_command != 0:
    command = input().split()
    quantity_command -= 1
    if command[0] == 'Insert':
        heap.append(int(command[1]))
        long += 1
        siftUp(heap, long - 1)
    elif command[0] == 'ExtractMax':
        HeapExtractMax(heap)
```

```
def mergesort(SP, ar1, ar2):
    temporary_sp = []
    global inverse
    count1, count2 = 0, 0
    len_ar1, len_ar2 = len(ar1), len(ar2)
    while count1 < len_ar1 and count2 < len_ar2:
        if ar1[count1] > ar2[count2]:
            inverse += len_ar1 - count1
            count2 += 1
        else:
            count1 += 1

    count1, count2 = 0, 0
    while count1 < len_ar1 and count2 < len_ar2:
        if ar1[count1] >= ar2[count2]:
            temporary_sp.append(ar2[count2])
            count2 += 1
        else:
            temporary_sp.append(ar1[count1])
            count1 += 1
    if count1 == len_ar1:
        temporary_sp.extend(ar2[count2:])
    else:
        temporary_sp.extend(ar1[count1:])
    SP.append(temporary_sp)


inverse = 0
len_sp = int(input())
sp = [[int(i)] for i in input().split()]
count = 0
sp2 = []
if len_sp % 2 == 1:
    sp.insert(0, [0])
    len_sp += 1
while len_sp > 1:
    while count < len_sp - 1:
        mergesort(sp2, sp[count], sp[count+1])
        count += 2
    sp = sp2
    sp2 = []
    len_sp //= 2
    if len_sp != 1 and len_sp % 2 == 1:
        sp.insert(0, [0])
        len_sp += 1
    count = 0

print(inverse)
```

```
def decision(ar, ln):
    ans = [0] * ln
    for i in range(1, ln):
        for j in range(i):
            if ar[i] % ar[j] == 0 and ans[j] + 1 > ans[i]:
                ans[i] = ans[j] + 1

    print(max(ans)+1)


def main():
    n = int(input())
    array = [int(i) for i in input().split()]
    decision(array, n)


if __name__ == "__main__":
    main()
```

```
def func(string_one, string_two):
    n = len(string_one)+1
    m = len(string_two)+1
    table = [[0 for i in range(m)] for i in range(n)]

    for i in range(1, n):
        table[i][0] = i
    for j in range(1, m):
        table[0][j] = j

    for i in range(1, n):
        for j in range(1, m):
            diff = 0 if string_one[i-1] == string_two[j-1] else 1
            table[i][j] = min(table[i-1][j]+1, 
                              table[i][j-1]+1, 
                              table[i-1][j-1]+diff)
            
    return print(table[-1][-1])


first = [i for i in input().split()]
second = [i for i in input().split()]

func(first, second)
```
